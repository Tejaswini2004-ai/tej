```python
# Parse date correctly
df = pd.read_csv('/content/dataset.csv', parse_dates=['Transaction_Date'])

# Handle missing values
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import pandas as pd

num_cols = df.select_dtypes(include='number').columns
imputer = KNNImputer(n_neighbors=5)
df[num_cols] = imputer.fit_transform(df[num_cols])

# Outlier Treatment
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df[col] = df[col].clip(Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)

# Scaling
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Feature Engineering (compute sales amount)
# Assuming 'Revenue' already includes discount; if not, you may multiply manually
df['Sales_Amount'] = df['Revenue']

# Calculate snapshot date for RFM
snapshot_date = df['Transaction_Date'].max() + pd.Timedelta(days=1)

# RFM Table
rfm = df.groupby('Customer_ID').agg({
    'Transaction_Date': lambda x: (snapshot_date - x.max()).days,
    'Sales_Amount': 'sum',
    'Transaction_Date': 'count'
}).rename(columns={
    'Transaction_Date': 'Recency',     # first lambda overrides the name, fix below
    'Sales_Amount': 'Monetary',
    'Transaction_Date': 'Frequency'
})

# Fix 'Recency' column
rfm['Recency'] = df.groupby('Customer_ID')['Transaction_Date'].max().apply(lambda x: (snapshot_date - x).days)
rfm['Frequency'] = df.groupby('Customer_ID')['Transaction_Date'].count()
rfm = rfm[['Recency', 'Frequency', 'Monetary']]

print(rfm.head())
```

```python
# Create churn label: e.g., no purchase in last 90 days
rfm['Churn'] = (rfm['Recency'] > 90).astype(int)

X = rfm[['Recency','Frequency','Monetary']]
y = rfm['Churn']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state=42)

clf = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, clf.predict_proba(X_test)[:,1]))
```
